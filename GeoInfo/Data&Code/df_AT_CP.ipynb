## Correlation between AT and CP

```python
import matplotlib.pyplot as plt
import numpy as np
from pandas import DataFrame
# create figure and axis objects with subplots()
fig,ax = plt.subplots(figsize=(8, 5))
# make a plot
ax.plot(df_CP_AT['year'], df_CP_AT['AT'], color="red", marker="o")
# set x-axis label
ax.set_xlabel("year",fontsize=14)
# set y-axis label
ax.set_ylabel("Average Temperature(AT)",color="red",fontsize=14)

ax.set_title('Relationship between AT and CP')

# twin object for two different y-axis on the sample plot
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(df_CP_AT['year'], df_CP_AT["CP"],color="blue",marker="o")
ax2.set_ylabel("Cumulative Precipitation(CP)",color="blue",fontsize=14)
plt.show()
# save the plot as a file
fig.savefig('F:/Masters/Semester2/GeoInfo/asessment/precipitation_extracted/AT_CP_Correlation.jpg',
            format='jpeg',
            dpi=100,
            bbox_inches='tight')
```
